```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
log = np.load('log.npz')
raw = log['raw']
scaled = log['scaled']
```

```python
np.max(np.abs(raw), axis=0)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
ax.plot(raw[:, 0], label='x')
ax.plot(raw[:, 1], label='y')
ax.plot(raw[:, 2], label='z')
#ax.plot(raw[:, 3], label='yaw')
ax.legend()
ax.set_xlabel('Time (s)')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
ax.plot(scaled[:, 0], label='x')
ax.plot(scaled[:, 1], label='y')
ax.plot(scaled[:, 2], label='z')
#ax.plot(raw[:, 3], label='yaw')
ax.legend()
ax.set_xlabel('Time (s)')
plt.show()
```

```python
# Integrate velocities to get a position trajectory
dt = 0.05
pos = np.zeros((scaled.shape[0], 3))
for i in range(1, scaled.shape[0]):
    pos[i] = pos[i-1] + scaled[i,:3] * dt

# Plot position trajectory
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
ax.plot(pos[:, 0], pos[:,1])
plt.axis('equal')
plt.show()
```
