```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define trajectory
RATE = 10    # hz 
DT = 1/RATE  # seconds
T = 10.0     # seconds
N = int(T/DT)  # number of points
tspan = np.linspace(0, T, N)

# Circle trajectory (centered at (1,0), radius 1)
x = 1 + np.cos(tspan)
y = np.sin(tspan)
traj_z = np.ones(N)
vx = 0.05 * -np.sin(tspan)
vy = 0.05 * np.cos(tspan)
vz = np.zeros(N)
vr = np.zeros(N)

# Plot
plt.plot(x, y)
plt.axis('equal')
# Plot quivers for vx and vy
for i in range(0, N, 1):
    plt.quiver(x[i], y[i], vx[i], vy[i])

plt.show()
```
